Beer-Lambert Law for absorbance as a function of distance, where $M$ is the molar concentration of the particles (in $mol/L$) and $\epsilon (\lambda)$ is the extinction coefficient (in $L/(mol\cdot cm$)) that is obtained from the measured absorption spectrum.
\begin{equation}
    p(\Delta s; \lambda) = 1 - 10^{-\epsilon(\lambda)M\Delta s}.
\end{equation}

The initial bounce-escape cone condition, where $l_z$ is the LSC height.
\begin{equation}
    \Delta s_0 = 2l_z.
\end{equation}


Subsequent reemission events after initial bounce-escape cone is satisfied, where $\xi \sim \mathcal{U}(0,1)$.
\begin{equation}
    \Delta s = -\frac{1}{\epsilon(\lambda)M}\log_{10} \xi.
\end{equation} 

The photon is absorbed after reemission with quantum yield $QY$,
\begin{equation}
    \xi < QY.
\end{equation}

The photon's new position is updated on incident direction with direction cosines $\mu$ as,
\begin{align}
    x' &= x + \mu_x \Delta s \\
    y' &= y + \mu_y \Delta s \\
    z' &= z + \mu_z \Delta s. 
\end{align}

For isotropic QDs, the new azimuthal angle and cosine of the deflection angle are obtained using 
\begin{align}
    \phi &= 2 \pi \xi \\
    \cos\theta &= sign(\chi)-\chi \\
    \chi &\equiv 2\xi - 1
\end{align}
so that $\phi\in [0,2\pi)$ and $\theta\in (0,\pi)$ and the emitted light is isotropically distributed. Using the two above equations, the direction cosines are updated according to
\begin{align}
    \mu_x' &= \frac{\sin\theta}{\sqrt{1-\mu_z^2}}(\mu_x\mu_z\cos\phi-\mu_y\sin\phi)+\mu_x\cos\theta\\
    \mu_y' &= \frac{\sin\theta}{\sqrt{1-\mu_z^2}}(\mu_y\mu_z\cos\phi+\mu_x\sin\phi)+\mu_y\cos\theta\\
    \mu_z' &= -\sin\theta\cos\phi\sqrt{1-\mu_z^2}+\mu_z\cos\theta
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection, Line3DCollection



def plot_lsc(lsc_definition,fig,ax):
    lsc_definition_array = [
        np.array(list(item))
        for item in lsc_definition
    ]

    points = []
    points += lsc_definition_array
    vectors = [
        lsc_definition_array[1] - lsc_definition_array[0],
        lsc_definition_array[2] - lsc_definition_array[0],
        lsc_definition_array[3] - lsc_definition_array[0]
    ]

    points += [lsc_definition_array[0] + vectors[0] + vectors[1]]
    points += [lsc_definition_array[0] + vectors[0] + vectors[2]]
    points += [lsc_definition_array[0] + vectors[1] + vectors[2]]
    points += [lsc_definition_array[0] + vectors[0] + vectors[1] + vectors[2]]

    points = np.array(points)

    edges = [
        [points[0], points[3], points[5], points[1]],
        [points[1], points[5], points[7], points[4]],
        [points[4], points[2], points[6], points[7]],
        [points[2], points[6], points[3], points[0]],
        [points[0], points[2], points[4], points[1]],
        [points[3], points[6], points[7], points[5]]
    ]

    faces = Poly3DCollection(edges, linewidths=1, edgecolors='k')
    faces.set_facecolor((0,0,1,0.1))
    
    ax.add_collection3d(faces)

    # Plot the points themselves to force the scaling of the axes
    ax.scatter(points[:,0], points[:,1], points[:,2], s=0)
    ax.set_aspect('equal')
```

Ray coordinate geometry.

Ray origin $P_0=(\frac{l_x}{2},\frac{l_y}{2},2l_z)$. 

Launch vector
\begin{align}
    \vec{v} &= \xi_1\vec{i}+\xi_2\vec{j}+\xi_3\vec{k}\\
    \lvert \vec{v} \rvert &= \sqrt{\xi_1^2+\xi_2^2+\xi_3^2} \\
    \hat{v} &= \frac{\vec{v}}{\lvert\vec{v}\rvert} = \frac{\xi_1\vec{i}+\xi_2\vec{j}-\xi_3\vec{k}}{\sqrt{\xi_1^2+\xi_2^2+\xi_3^2}}
\end{align}
where $\xi_i \sim \mathcal{U}(0,1)$

Equation of the line passing through $P_0$ in direction $\hat{v}$
\begin{align}
    \vec{OP_0} &= \frac{l_x}{2}\vec{i} + \frac{l_y}{2}\vec{j} + 2l_z\vec{k} \\
    \mathcal{L} &= {OP_0}+\hat{v}\cdot t \\
    &= (\frac{l_x}{2}+\frac{\xi_1}{\lvert \vec{v} \rvert)}\cdot t)\vec{i}+(\frac{l_y}{2}+\frac{\xi_2}{\lvert \vec{v} \rvert)}\cdot t)\vec{j}+(2l_z -\frac{\xi_3}{\lvert \vec{v} \rvert)}\cdot t)\vec{k}
\end{align}
where $t$ is a scalar. More generally,
\begin{align}
    \mathcal{L} &= (x_0+\frac{\xi_1}{\lvert \vec{v} \rvert)}\cdot t)\vec{i}+(y_0+\frac{\xi_2}{\lvert \vec{v} \rvert)}\cdot t)\vec{j}+(z_0-\frac{\xi_3}{\lvert \vec{v} \rvert)}\cdot t)\vec{k}
\end{align}

Plane of intersection: $z=l_z$ subject to constraint of $0\leq x\leq l_x$ and $0\leq y \leq l_y$.
\begin{align}
    (z_0-\frac{\xi_3}{\lvert \vec{v} \rvert)}\cdot t) = l_z \\
    t = \frac{\lvert\vec{v}\rvert}{\xi_3}(z_0-l_z)
\end{align}

Applying this to the constraint, the bounds for the random variable in the x- and y- directions can be obtained.
\begin{align}
    0 \leq x_0+\frac{\xi_1}{\lvert \vec{v} \rvert)}\cdot t &\leq l_x \\
    x_0+\frac{\xi_1}{\lvert \vec{v} \rvert)}\cdot t &\geq 0 \\
    x_0 &\geq -\frac{\xi_1}{\lvert \vec{v} \rvert)}\cdot \frac{\lvert\vec{v}\rvert}{\xi_3}(z_0-l_z)\\
    x_0 &\geq -\frac{\xi_1}{\xi_3}(z_0-l_z)\\
    x_0+\frac{\xi_1}{\lvert \vec{v} \rvert)}\cdot t &\leq l_x \\
    x_0+\frac{\xi_1}{\xi_3}(z_0-l_z) &\leq l_x \\
    x_0 &\leq l_x -\frac{\xi_1}{\xi_3}(z_0-l_z) \\
    -\frac{\xi_1}{\xi_3}(z_0-l_z)\leq x_0 &\leq l_x -\frac{\xi_1}{\xi_3}(z_0-l_z) 
\end{align}
Similarly for $y_0$,
\begin{align}
    -\frac{\xi_2}{\xi_3}(z_0-l_z)\leq y_0 &\leq l_y -\frac{\xi_2}{\xi_3}(z_0-l_z) 
\end{align}

Or, as an alternative method, place the photon on the LSC and trace its ray back to the boundary. $\theta$ and $\gamma$ denote the incident and azimuthal angles respectively. Minimise $\lvert t \rvert$ in
\begin{align}
    \vec{OP_i}+t\vec{v}
\end{align}

Case 1, $x = 0 || x = l_x$
\begin{align}
    (x_i &+ t\cos\gamma\sin\theta = 0)\, ||\, (x_i + t\cos\gamma\sin\theta = l_x)\\
    \lvert t \rvert &= \left| \frac{-x_i}{\cos\gamma\sin\theta} \right|, \left|\frac{l_x-x_i}{\cos\gamma\sin\theta} \right|
\end{align}

Case 2, similarly
\begin{align}
    (y_i &+ t\sin\gamma\sin\theta = 0)\, ||\, (y_i + t\sin\gamma\sin\theta = l_y)\\
    \lvert t \rvert &= \left| \frac{-y_i}{\sin\gamma\sin\theta} \right|, \left|\frac{l_y-y_i}{\sin\gamma\sin\theta} \right|
\end{align}

Case 3, $z=1$
\begin{align}
    z_i &+ t \cos\theta = 1 \\
    \lvert t \rvert &= \left| \frac{1 - z_i}{\cos\theta} \right|
\end{align}

```python
def launch_photon(lsc_dimensions):
    
    # LSC Dimensions
    l_x = float(lsc_dimensions[0])
    l_y = float(lsc_dimensions[1])
    l_z = float(lsc_dimensions[2])
    
    refire = True
    if refire:
        # Random launch variables, cartesian coordinates of the photon, 
        # plus azimuthal and incident angle 
        x_i = np.random.random_sample()*l_x
        y_i = np.random.random_sample()*l_y
        z_i = l_z
        gamma_i = np.random.random()*2*np.pi # azimuthal angle
        theta_i = 0.501*np.pi+np.random.random()*np.pi # incident angle

        # Ray max values
        x_max = 1.2*l_x
        y_max = 1.2*l_y
        z_max = 2.0*l_z

        # Trace vector back up to boundary
        t_s = [0]*5 # 5 Values, three cases
        # At x boundary
        t_s[0] = x_i/(np.cos(gamma_i)*np.sin(theta_i))
        t_s[1] = (x_max-x_i)/(np.cos(gamma_i)*np.sin(theta_i))
        t_s[2] = y_i/(np.sin(gamma_i)*np.sin(theta_i))
        t_s[3] = (y_max-y_i)/(np.sin(gamma_i)*np.sin(theta_i))
        t_s[4] = (z_max-z_i)/(np.cos(theta_i))
        # Get smallest negative value of t
        array = np.array(t_s)
        t_intersect_idx = list(array).index(max(array[array<0]))
        x_0 = x_i + np.cos(gamma_i)*np.sin(theta_i)*t_s[t_intersect_idx]
        y_0 = y_i + np.sin(gamma_i)*np.sin(theta_i)*t_s[t_intersect_idx]
        z_0 = z_i + np.cos(theta_i)*t_s[t_intersect_idx]

        # Exit to avoid skewing the domain
        if not(x_0 < -0.2*l_x):
            refire = False
        if not(y_0 < -0.2*l_y):
            refire = False
    
    incident_ray_p0 = np.array([x_0,y_0,z_0])
    incident_ray_pi = np.array([x_i,y_i,z_i])
    incident_ray_points = [incident_ray_p0,incident_ray_pi]
    
    # Determine incident ray as a vector, normalised
    v_xi = x_i - x_0
    v_yi = y_i - y_0
    v_zi = z_i - z_0
    incident_ray_vector = np.array([v_xi,v_yi,v_zi])/np.linalg.norm([v_xi,v_yi,v_zi])
    incident_ray_angles = [theta_i,gamma_i]

    # Return point on LSC, vector and incident/azimuthal angles for later
    return (incident_ray_points, incident_ray_vector, incident_ray_angles)
    
def plot_incident_ray(lsc_dimensions,incident_ray_points,ax):  
    
    p0 = incident_ray_points[0]
    pi = incident_ray_points[1]
    # Plot line between these two points
    x_s = [p0[0], pi[0]]
    y_s = [p0[1], pi[1]]
    z_s = [p0[2], pi[2]]
    ax.plot(x_s,y_s,z_s,color='red') 
    
```

The incident vector $\vec{i}$, reflected vector $\vec{r}$, the refracted vector $\vec{t}$ and the vector that is orthogonal to the surface $\vec{n}$ are normalised.
\begin{equation}
    \lvert\vec{i}\rvert = \lvert\vec{r}\rvert = \lvert\vec{t}\rvert = \lvert\vec{n}\rvert = 1
\end{equation}
Any vector can be expressed in terms of vector components that are orthogonal and tangential to the LSC surface. In this case, the orthogonal component, $\vec{v_{\perp}}$ would be in the direction of the normal vector $\vec{n}$. The tangential component would be in the $x-y$ plane, $\vec{v_{\parallel}}$.
\begin{align}
    \vec{v_{\perp}} &= \frac{\vec{v}\cdot\vec{n}}{\lvert\vec{n}\rvert^2}\vec{n}=(\vec{v}\cdot\vec{n})\vec{n} \\
    \vec{v_{\parallel}} &= \vec{v}-\vec{v_{\perp}} \\
    \cos\theta_v &= \frac{\lvert\vec{v_{\perp}}\rvert}{\lvert\vec{v}\rvert} = \lvert\vec{v_{\perp}}\rvert\\
    \rightarrow \cos\theta_v &= \pm\vec{v}\cdot\vec{n} \\
    \sin\theta_v &= \frac{\lvert\vec{v_{\parallel}}\rvert}{\lvert\vec{v}\rvert} = \lvert\vec{v_{\parallel}}\rvert
\end{align}


For reflection $\theta_r=\theta_i$, applied to above this means that
\begin{align}
    \lvert\vec{r_{\perp}}\rvert &= \cos\theta_r=\cos\theta_i=\lvert\vec{i_{\perp}}\rvert \\
    \lvert\vec{r_{\parallel}}\rvert &= \sin\theta_r=\sin\theta_i=\lvert\vec{i_{\parallel}}\rvert \\
    \vec{r_{\perp}} &= -\vec{i_{\perp}} \\
    \vec{r_{\parallel}} &= \vec{i_{\parallel}} \\
    So,\,\vec{r} &= \vec{r_{\perp}} + \vec{r_{\parallel}} = \vec{i_{\parallel}} -\vec{i_{\perp}} 
\end{align}
Finally, the reflection vector can be expressed using just $\vec{i}$ and $\vec{n}$.
\begin{align}
    \vec{r} &=  \vec{i_{\parallel}} -\vec{i_{\perp}} \\
    &= [\vec{i}-(\vec{i}\cdot\vec{n})\vec{n}] - (\vec{i}\cdot\vec{n}) \\
    &= \vec{i}-2(\vec{i}\cdot\vec{n})\vec{n}
\end{align}


Refraction is a little bit trickier. Starting with Snell's Law.
\begin{align}
    n_1\sin\theta_i &= n_2\sin\theta_t \rightarrow \sin\theta_t = \frac{n_1}{n_2}\sin\theta_i \\
    \sin\theta &\leq 1\therefore\sin\theta_i\geq\frac{n_2}{n_1}
\end{align}
A similar method is used for the refraction or transmission vector,
\begin{align}
    \vec{t} = \vec{t_{\perp}}+\vec{t_{\parallel}}
\end{align}
and from the equation for $\sin\theta_v$ earlier, the $\sin\theta$ values can be subsituted for the parallel vectors. Also they are parallel and in the same direction.
\begin{align}
    \lvert\vec{t_{\parallel}}\rvert &= \frac{n_1}{n_2}\lvert\vec{i_{\parallel}}\rvert \\
    &= \frac{n_1}{n_2}[\vec{i} + \cos\theta_i \vec{n}]
\end{align}
For the orthogonal component, Pythagoras is used. Then, the orthogonal and parallel components are incorporated into the equation for $\vec{t}$ and simplification is done to obtain an equation in $\vec{i}$ and $\vec{n}$.
\begin{align}
    \vec{t}_{\perp} &= -\sqrt{1-\lvert\vec{t}_{\parallel}\rvert^2}\vec{n} \\
    \vec{t} &= \frac{n_1}{n_2}\vec{i} + \Big(\cos\theta_i -\sqrt{1-\lvert\vec{t}_{\parallel}\rvert^2}\Big)\vec{n} \\
    &= \frac{n_1}{n_2}\vec{i} + \Big(\cos\theta_i -\sqrt{1-\sin^2\theta_t}\Big)\vec{n}
\end{align}
The $\sin^2\theta_t$ term can be calculated by Snell's Law.
\begin{equation}
    \sin^2\theta_t = \Big(\frac{n_1}{n_2}\Big)^2\sin^2\theta_i=\Big(\frac{n_1}{n_2}\Big)^2(1-\cos^2\theta_i)
\end{equation}

The critical angle appears because the above equation can only work for non-negative values under the square root. This gives a new condition for the refracted vector, which cannot be found if the condition is violated. It is also necessary that this provides a critical angle, $\theta_c$, for which the refracted angle can only be found if the incident angle, $\theta_i$ is less than this criterion.

\begin{align}
    \sin^2\theta_t \leq 1 \\
    \theta_c = \arcsin\frac{n_2}{n_1}\leftrightarrow n_1 > n_2
\end{align}

The Fresnel equations apply to light as an electromagnetic wave in which each photon that is incident on an interface between two different media is reflected and/or transmitted. This ratio is decribed for the orthogonal and parallel polarisations of light as a function of $n_1$, $n_2$, $\theta_i$ and $\theta_t$.
\begin{align}
    R_{\perp}(\theta_i) &= \Bigg(\frac{n_1\cos\theta_i-n_2\cos\theta_t}{n_1\cos\theta_i+n_2\cos\theta_t}\Bigg)^2\\
    R_{\parallel}(\theta_i) &= \Bigg(\frac{n_2\cos\theta_i-n_1\cos\theta_t}{n_2\cos\theta_i+n_1\cos\theta_t}\Bigg)^2
\end{align}

Assuming that the incoming light is unpolarised then the reflectance $R(\theta_i)$ and transmittance $T(\theta_i)$ as a function of incident angle $\theta_i$ can be found by taking the average of both polarisations in the Fresnel equations.

\begin{equation}
  R(\theta_i) = 
  \begin{cases}
      \frac{R_{\perp}(\theta_i)+R_{\parallel}(\theta_i)}{2} &\leftrightarrow \neg TIR \\
      1  &\leftrightarrow TIR
  \end{cases}
\end{equation}

The transmission can be determined from the relation of $R+T=1$.
\begin{equation}
    T(\theta_i) = 1 - R(\theta_i)
\end{equation}
